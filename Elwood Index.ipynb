## Creating function for streaming data

### Average trade weighted price 1min

```python
def weighted_price(ob,size_required,rounding):
    prices = {}
    sizes = {}
    prices["bid"] = [x["bid_price"] for x in ob]
    sizes["bid"] = [x["bid_size"] for x in ob]
    prices["ask"] = [x["ask_price"] for x in ob]
    sizes["ask"] = [x["ask_size"] for x in ob]
    weighted_avg = {"bid":0,"ask":np.inf}
    for side in ["bid","ask"]:
        if sizes[side][0] >= size_required:
            weighted_avg[side] = prices[side][0]
        elif sum(sizes[side]) >= size_required:
            found_size = 0
            for level in range(len(prices[side])):
                found_size += sizes[side][level]
                if found_size >= size_required:
                    depth = level
                    break
            ### find average of depth-1 , then add amaount required from last level
            avg_price = 0
            for x in range(depth):
                avg_price += sizes[side][x] * prices[side][x]
            tot_sizes = sum(sizes[side][:depth])
            size_on_last_level = size_required-tot_sizes
            avg_price += size_on_last_level*prices[side][depth]
            ### calc avg ###
            weighted_avg[side] = round(avg_price/size_required/rounding,0)*rounding
        else:
            print("20 levels do not have enough size to trade a "+side)
    return weighted_avg

bit = bitstamp_snapshots.head(20).to_dict('records')
bid_ask_weights = weighted_price(bit,2000,0.25)
print(bid_ask_weights)
print(np.mean([bid_ask_weights["bid"],bid_ask_weights["ask"]]))
```

### Create timestamps for buckets

```python
## the "time_length" MUST be a factor of the whole "time_interval" for it to work 
## return rounded down timestamp and the next timebucket stamp as an array
def round_timestamp (timestamp, time_interval='minutes',time_length=1):
    timestamp_breakdown = {
        'years' : timestamp.year,
        'months' : timestamp.month,
        'days' : timestamp.day,
        'hours' : timestamp.hour,
        'minutes' : timestamp.minute,
        'seconds' : timestamp.second,}
    
    times = ['years','months','days','hours','minutes','seconds']
    idx = times.index(time_interval)+1
    reset_interval = times[idx:len(times)]
    
    rounded_down = int(timestamp_breakdown[time_interval] / time_length) * time_length
    timestamp_breakdown[time_interval] = rounded_down
    for interval in reset_interval:
        timestamp_breakdown[interval] = 0
        
    t = timestamp_breakdown.copy()
    start_timestamp = datetime.datetime(year=t["years"],month=t["months"],day=t["days"],hour=t["hours"],minute=t["minutes"],second=t["seconds"])

    if time_interval == 'seconds':
        end_timestamp = start_timestamp+datetime.timedelta(seconds=time_length)
    elif time_interval == 'minutes':
        end_timestamp = start_timestamp+datetime.timedelta(minutes=time_length)
    return start_timestamp, end_timestamp

round_timestamp(datetime.datetime.now(),time_interval='minutes',time_length=1)
```

```python
def interval_price(df_trades, df_ob, required_size, time_length, time_interval='minutes', rounding=0.01):

    df_trades["timestamp"] = pd.to_datetime(df_trades["timestamp"])
    
    # initialise
    total_price = 0
    total_quantity = 0
    buckets_avg = []
    test = []
    missed_timestamps = []
    first = True
    
    ## acts as a data streamer for trades (is looping through a dataframe)
    for index in range(len(df_trades)):
        row = df_trades.iloc[index]
        timestamp = pd.to_datetime(row["timestamp"])
        if first == True:
            start_timestamp, end_timestamp = round_timestamp(timestamp,time_interval=time_interval,time_length=time_length)
            first = False
            
        # if the timestamp jumps into the next bucket, then finish current bucket
        if timestamp > end_timestamp:
            weighted_avg = total_price/ total_quantity
            weighted_avg_round = round(round(weighted_avg / rounding) * rounding,10)
            buckets_avg.append((start_timestamp,weighted_avg_round))
            total_price = 0
            total_quantity = 0  
            test=[]
            
            while True:
                start_timestamp = end_timestamp
                end_timestamp = end_timestamp + datetime.timedelta(minutes=time_length)
                if timestamp > end_timestamp:
                    missed_timestamps.append(start_timestamp)
                    
                    ### create mid average based on orderbook snapshot
                    snap_shot = df_ob[(df_ob.index>=start_timestamp) & (df_ob.index<end_timestamp)].to_dict('records')
                    bid_ask_weights = weighted_price(snap_shot,required_size,0.00000001)
                    weighted_avg = np.mean((bid_ask_weights["bid"],bid_ask_weights["ask"]))
                    weighted_avg_round = round(round(weighted_avg / rounding) * rounding,10)
                    buckets_avg.append((start_timestamp,weighted_avg_round))
                else:
                    break
                        
        # if timestamp in current bucket then create weighted average price
        total_price += (row["qty"] * row["price"])
        total_quantity += row["qty"]  
        test.append((row["qty"],row["price"]))
    return buckets_avg # missed_timestamps
```

## 1 minute buckets - weighted average price
Trades: Price x Size / Size <br>
Else, exchange 1BTC weighted mid price <br>
Weighted price for 1BTC on ask and bid / 2. Levels may be truncacted to get exactly 1BTC. <br>
If not enough size to reach 1BTC, then take the average of what is availabe <br>

```python
coinbase_buckets = interval_price(coinbase_trades, coinbase_snapshots, required_size=300, time_length=1)
bitstamp_buckets = interval_price(bitstamp_trades, bitstamp_snapshots, required_size=300, time_length=1)
coinbase_buckets
```

## 1 hour fixing window 
Simple moving average of the last 60 interval prices (1min)

```python
coinbase_1h = round(np.mean([bucket[1] for bucket in coinbase_buckets]),2)
bitstamp_1h = round(np.mean([bucket[1] for bucket in bitstamp_buckets]),2)
print("coinbase:",coinbase_1h,"bitstamp:",bitstamp_1h)
```

## Elwood hourly reference rate
each Exchange price * weight / Total weights

```python
def reference_rate(exch_one, weight_one, exch_two, weight_two):
    return ((exch_one * weight_one) + (exch_two * weight_two)) / (weight_one + weight_two)

reference_rate (coinbase_1h, 0.5, bitstamp_1h, 0.5)
```
